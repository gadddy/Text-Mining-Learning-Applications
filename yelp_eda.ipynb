# Generate wordcloud for Reviews

```python
import re
import nltk
from nltk.corpus import PlaintextCorpusReader

# Combine all text into a single string
text = " ".join(df['review_text'].tolist())

# Split the string into words and create the corpus
reviews =  text.split()


#1. First change every token to lower case.
reviews_lower = [w.lower() for w in reviews]

#2.Use regular expressions to keep only alphabetic tokens.

reviews_lower_use = [w for w in reviews_lower if re.search('^[a-z]+$',w)]
print(reviews_lower_use[0:30])

#3. Find out the most frequent tokens after these steps

fdist_reviews = nltk.FreqDist(reviews_lower_use)
print(fdist_reviews.most_common(10))
```

```python
#remove stopwords
from nltk.corpus import stopwords
stop_list = stopwords.words('english')
reviews_stopremoved = [w for w in reviews_lower_use if w not in stop_list]
fdist_reviews = nltk.FreqDist(reviews_stopremoved)
print(reviews_stopremoved[0:30])
```

```python
#create wordcloud
import wordcloud
import matplotlib.pyplot as plt

# Create and display the word cloud
wc = wordcloud.WordCloud(background_color='white',width=800, height=600).generate_from_frequencies(fdist_reviews)
wc.to_file('review_wc.png')
plt.imshow(wc)
plt.axis("off")
plt.show()
```

```python
#ATTEMPT TO GET MISSPELLINGS EXAMPLES, is very barebone and inaccurate
from spellchecker import SpellChecker
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

count = 0

# Load English stop words
stop_words = set(stopwords.words('english'))

# Initialize spell checker
spell = SpellChecker()

# Array to hold strings with spelling errors
errors = []

misspelled_words = []

# Function to check spelling in a string
def remove_punctuation(words):
    words_no_punct = [word.translate(str.maketrans('', '', string.punctuation)) for word in words]
    return words_no_punct


def find_spelling_errors(sentence):
    # Tokenize the sentence into words
    words = sentence.split()
    # Filter out stop words

    filtered_words = [word for word in words if word.lower() not in stop_words]
    # Remove punctuation from the remaining words
    #words_no_punct = [word.translate(str.maketrans('', '', string.punctuation)) for word in filtered_words]
    # Find misspelled words
    words_no_punct = remove_punctuation(filtered_words)
    filtered_words = [word for word in words_no_punct if word.lower() not in stop_words]
    misspelled = spell.unknown(filtered_words)
    return list(misspelled)

# Iterate over DataFrame's 'text' column
for text in df["review_text"].unique():
    temp = {}
    misspelled = find_spelling_errors(text)
    if misspelled:  # If there are misspelled words
        
        temp['text'] = text  # Append the entire sentence to errors
        misspelled_words.extend(misspelled)  # Append individual misspelled words
        temp['spell errors'] = misspelled
        count += 1
        errors.append(temp)
    if count ==50:
        break

for e in errors:
    print("---")
    print(e)

print(misspelled_words)
```

```python
import string  
def split_into_word_and_punctuation_suffix(word):
    for i, n in list(enumerate(word))[::-1]:
        if n not in string.punctuation:
            return word[:i+1], word[i+1:]
    return "", word

def fix_misspelt_word(word):
    word, punct = split_into_word_and_punctuation_suffix(word)
    #print(word," ", punct)
    if spell.correction(word) is not None:
        return spell.correction(word) + punct
    return word + punct

def fix_misspelt_sentence(sentence):
    return " ".join([fix_misspelt_word(word) for word in sentence.split()])

fix_misspelt_sentence("Yet again.. Disssappointed. Now they don't have bruschetta. Really bad deal.")
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import gensim
from gensim import corpora


# Sample documents
documents = [
  "Machine learning is a field of computer science that uses statistical methods to give computers the ability to learn without being explicitly programmed.",
  "Deep learning is a type of machine learning that uses artificial neural networks, inspired by the structure and function of the brain, to learn patterns from data.",
  "Natural language processing (NLP) is a subfield of computer science, information engineering, and artificial intelligence concerned with the interactions between computers and human language."
]

# Load English stop words
stop_words = set(stopwords.words('english'))


# Preprocess text (remove stopwords, stemming etc.)
preprocessed_documents = [[w.lower() for w in word_tokenize(text) if w.lower() not in stop_words] for text in documents]

# Create dictionary
dictionary = corpora.Dictionary(preprocessed_documents)

# Create document-term matrix
corpus = [dictionary.doc2bow(doc) for doc in preprocessed_documents]

# Train LDA model
lda_model = gensim.models.LdaMulticore(corpus, id2word=dictionary, num_topics=5)

# Print topics
for idx, topic in lda_model.print_topics(num_words=3):
  print(f"Topic {idx}:", topic)
```
