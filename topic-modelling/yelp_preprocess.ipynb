```python
import pandas as pd
unique_df = pd.read_csv('unique_data.csv')
```

```python
import tqdm
#for each 'review_text' row, count the number of ' in text and get highest count row
unique_df['num_apostrophes'] = unique_df['review_text'].apply(lambda x: x.count("'"))
```

```python
#generate character length of each 'review_text' row
unique_df['char_length'] = unique_df['review_text'].apply(lambda x: len(x))
```

```python
#create new df with just 'review_stars', 'review_text', 'num_apostrophes', 'char_length'   columns
new_df = unique_df[['review_stars', 'review_text', 'num_apostrophes', 'char_length']]
```

```python
new_df.to_csv('unique_data_count.csv', index=False)
```

```python
unique_df['review_stars'].value_counts()
```

```python
# Create distribution of 'review_stars' column
import matplotlib.pyplot as plt
import numpy as np

# Count the occurrences of each unique value
counts = unique_df['review_stars'].value_counts(normalize=True) * 100

# Sort by index for ordered bars
counts = counts.sort_index()

# Create a bar chart
ax = counts.plot(kind='bar')

plt.xlabel('Review Stars')
plt.ylabel('Percentage')
plt.xticks(rotation=0)

# Label each bar with its height
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()

#save figure
plt.savefig('review_stars_distribution.png')
```

```python
import math

# Get numeric columns
numeric_cols = unique_df.select_dtypes(include=np.number).columns

# Calculate the number of rows and columns for the grid
n = len(numeric_cols)
ncols = 3  # Number of columns
nrows = math.ceil(n / ncols)  # Number of rows

# Create a grid of subplots
fig, axs = plt.subplots(nrows, ncols, figsize=(15, nrows*5))

# Flatten the array of axes
axs = axs.flatten()

# Plot each histogram in a separate subplot
for i, col in enumerate(numeric_cols):
    unique_df[col].plot(kind='hist', title=col, ax=axs[i])

# Remove unused subplots
for i in range(n, nrows*ncols):
    fig.delaxes(axs[i])

plt.tight_layout()
plt.show()
```
