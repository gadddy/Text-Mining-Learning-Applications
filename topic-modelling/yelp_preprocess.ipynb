```python
import preprocess_df
df_set = unique_df


```

```python
corpus = preprocess_df.load_corpus(df_set, 'review_text')
docs = preprocess_df.corpus2docs(corpus)
```

```python
import gensim
dictionary = gensim.corpora.Dictionary(docs)
vecs = preprocess_df.docs2vecs(docs, dictionary)
```

```python
#I used 10 topics. You can try other numbers as well
lda = gensim.models.ldamodel.LdaModel(corpus=vecs, id2word=dictionary, num_topics=10)
```

```python
topics = lda.show_topics(10, 5)

for i in range(0, 10):
    print(topics[i])
```
